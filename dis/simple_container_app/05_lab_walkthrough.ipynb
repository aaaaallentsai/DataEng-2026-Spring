````python
# Example: What does data cleaning look like?
import pandas as pd

# Simulate some messy data
messy_data = pd.DataFrame({
    'event': ['  Dancing Plague  ', 'dancing plague', 'DANCING PLAGUE'],
    'location': ['Strasbourg', 'strasbourg', 'Strasbourg, France'],
    'year': [1518, None, 1518]
})

print("BEFORE cleaning:")
print(messy_data)
print()

# Clean the data - STEP BY STEP
cleaned = messy_data.copy()

# Step 1: Standardize text (strip whitespace, fix capitalization)
cleaned['event'] = cleaned['event'].str.strip().str.lower().str.title()
cleaned['location'] = cleaned['location'].str.strip().str.title()

# Step 2: Remove rows with missing critical data
cleaned = cleaned.dropna()  

# Step 3: Remove exact duplicates
cleaned = cleaned.drop_duplicates()

print("AFTER basic cleaning:")
print(cleaned)
print()

# ğŸ’¡ Notice: We still have 2 rows! Why?
# "Strasbourg" vs "Strasbourg, France" are technically different strings
# More advanced cleaning would normalize location names:
# - Extract just the city name
# - Use a geocoding service to standardize
# - Apply domain knowledge

print("âš ï¸  Real-world lesson: Basic deduplication isn't always enough!")
print("    'Strasbourg' â‰  'Strasbourg, France' (as strings)")
print("    You'd need additional logic to catch these as duplicates.")
````

### ğŸ”‘ Key Concepts Explained

#### 1. Base Image Selection
```dockerfile
FROM python:3.12-slim
```
**What is this?**
- Pre-built image with Python already installed
- `slim` = smaller size (good for simple apps)
- Built on Debian Linux â†’ uses `apt-get`

**Common mistake:** Using an incompatible base image

#### 2. Layer Caching Optimization
```dockerfile
COPY requirements.txt ...    # â† Copy deps FIRST
RUN pip install ...          # â† Install packages
COPY app.py ...             # â† Copy code LAST
```

**Why this order?**
- Each instruction = one layer
- Docker caches unchanged layers
- If `app.py` changes, packages don't reinstall!

**Example:**
```
Change app.py â†’ Only steps after COPY app.py rebuild
Change requirements.txt â†’ Everything from pip install rebuilds
```

#### 3. Port Exposure
```dockerfile
EXPOSE 8080
```

**This is documentation only!** It doesn't actually open the port.

Real port opening happens with `-p` flag:
```bash
docker run -p 8080:8080 myapp
             â†‘    â†‘
             |    â””â”€ Container port (8080)
             â””â”€â”€â”€â”€â”€â”€ Host port (8080)
```

#### 4. Wrong Base Image Pitfall
```dockerfile
# âŒ WRONG - Fedora uses dnf, not apt-get
FROM fedora:latest
RUN apt-get install python3  # Error!

# âœ… CORRECT
FROM python:3.12-slim
# No need to install Python - it's already there!
```

## Part 5: Web Scraping Introduction (7 minutes)

### ğŸ•¸ï¸ What is Web Scraping?

**Extracting structured data from websites** using code instead of manual copy-paste.

### ğŸ› ï¸ Tool: BeautifulSoup

BeautifulSoup parses HTML and lets you navigate it like a tree.

````python
# Simple BeautifulSoup Example
from bs4 import BeautifulSoup

# Sample HTML
html = """
<html>
  <body>
    <h1 class="title">Welcome to Web Scraping</h1>
    <p id="intro">This is an introduction.</p>
    <ul>
      <li>Python</li>
      <li>BeautifulSoup</li>
      <li>Requests</li>
    </ul>
  </body>
</html>
"""

# Parse the HTML
soup = BeautifulSoup(html, 'html.parser')

# Find elements
title = soup.find('h1', class_='title')
print(f"Title: {title.text}")

intro = soup.find('p', id='intro')
print(f"Intro: {intro.text}")

# Find all list items
items = soup.find_all('li')
print(f"\nList items:")
for item in items:
    print(f"  - {item.text}")
````

### ğŸ¯ Common BeautifulSoup Methods

| Method | Description | Example |
|--------|-------------|----------|
| `find()` | Find first matching element | `soup.find('h1')` |
| `find_all()` | Find all matching elements | `soup.find_all('p')` |
| `.text` | Get text content | `element.text` |
| `.get('attr')` | Get attribute value | `link.get('href')` |
| `select()` | CSS selector | `soup.select('.class')` |

````python
# More realistic example: Scraping links
html_with_links = """
<html>
  <body>
    <div class="nav">
      <a href="/home">Home</a>
      <a href="/about">About</a>
      <a href="/contact">Contact</a>
    </div>
  </body>
</html>
"""

soup = BeautifulSoup(html_with_links, 'html.parser')

# Find all links
links = soup.find_all('a')

print("Found links:")
for link in links:
    text = link.text
    href = link.get('href')
    print(f"  {text}: {href}")
````

### âœ… Ethical Scraping Guidelines

1. **Check robots.txt** - Respect site's scraping rules
2. **Rate limiting** - Don't hammer servers with requests
3. **Terms of Service** - Read and follow them
4. **Use APIs when available** - Often better than scraping
5. **Add delays** - `time.sleep()` between requests
